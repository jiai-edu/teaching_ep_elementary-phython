データ型dtype 型コード	説明\
int8	i1	符号あり8ビット整数型\
int16	i2	符号あり16ビット整数型\
int32	i4	符号あり32ビット整数型\
int64	i8	符号あり64ビット整数型\
uint8	u1	符号なし8ビット整数型\
uint16	u2	符号なし16ビット整数型\
uint32	u4	符号なし32ビット整数型\
uint64	u8	符号なし64ビット整数型\
float16	f2	半精度浮動小数点型（符号部1ビット、指数部5ビット、仮数部10ビット）\
float32	f4	単精度浮動小数点型（符号部1ビット、指数部8ビット、仮数部23ビット）\
float64	f8	倍精度浮動小数点型（符号部1ビット、指数部11ビット、仮数部52ビット）\
float128	f16	四倍精度浮動小数点型（符号部1ビット、指数部15ビット、仮数部112ビット）\
complex64	c8	複素数（実部・虚部がそれぞれfloat32）\
complex128	c16	複素数（実部・虚部がそれぞれfloat64）\
complex256	c32	複素数（実部・虚部がそれぞれfloat128）\
bool	?	ブール型（True or False）\
object	O	Pythonオブジェクト型\

```
df = pd.DataFrame(data=[[1, 2, 3], [4, 5, 6]], columns=['a', 'b', 'c'])
print(df)
print("\n")

a_values = df.values
print(a_values)
```

```
#valuesはビューを返し、一方の値を変更するともう一方の値も変更される

print(np.shares_memory(a_values, df))
print("\n")

a_values[0, 0] = 100
print(a_values)
print("\n")

print(df)
```

```
#常にビューを返すわけではなく、例えば各列のデータ型dtypeが異なる場合はコピー

df_if = pd.DataFrame(data=[[1, 0.1], [2, 0.2]], columns=['int', 'float'])
print(df_if)
print("\n")

print(df_if.dtypes)
print("\n")

a_values_if = df_if.values
print(a_values_if)
print("\n")

print(np.shares_memory(a_values_if, df_if))
print("\n")

a_values_if[0, 0] = 100
print(a_values_if)
print("\n")

print(df_if)
```

```
#行や列などの範囲を指定してvaluesを取得する場合は指定方法によってビューかコピーかが異なったりする

print(df[['a', 'c']].values)
print("\n")

print(np.shares_memory(df[['a', 'c']].values, df))
print("\n")

print(df.iloc[:, ::2].values)
print("\n")

print(np.shares_memory(df.iloc[:, ::2].values, df))
```

```
#ビューだと問題がある場合はvaluesのcopy()で明示的にコピーを生成すればよい

a_values_copy = df.values.copy()
print(a_values_copy)
print("\n")

print(np.shares_memory(a_values_copy, df))
print("\n")

a_values_copy[0, 0] = 10
print(a_values_copy)
print("\n")

print(df)
```

```
#コンストラクタpandas.DataFrame(), pandas.Series()にnumpy.ndarrayを渡すとメモリが共有される。
#一方の値を変更するともう一方の値も変更される

a = np.array([[1, 2, 3], [4, 5, 6]])
print(a)
print("\n")

df_a = pd.DataFrame(a, columns=['a', 'b', 'c'])
print(df_a)
print("\n")

print(np.shares_memory(a, df_a))
print("\n")

a[0, 0] = 100
print(a)
print("\n")

print(df_a)
print("\n")

df_a.iat[1, 0] = 10
print(df_a)
print("\n")

print(a)
```

```
#メモリを共有させたくない場合はnumpy.ndarrayのcopy()でコピーをコンストラクタの引数に指定する

df_a_copy = pd.DataFrame(a.copy(), columns=['a', 'b', 'c'])
print(df_a_copy)
print("\n")

a[0, 0] = 1
print(a)
print("\n")

print(df_a_copy)
```

```
#四捨五入ではなく偶数への丸め

from decimal import Decimal, ROUND_HALF_UP, ROUND_HALF_EVEN
import platform

print(platform.python_version())
print(pd.__version__)
print(np.__version__)
print("\n")

s_f = pd.Series([123.456, 654.123])
print(s_f)
print("\n")

#round()の引数に整数nを渡すと、小数点以下n桁に丸められる。
print(s_f.round())
print("\n")

print(s_f.round().astype(int))
```

```
#正の整数は小数点以下の桁数指定、負の整数は整数の桁数指定になる

print(s_f.round(2))
print("\n")

print(s_f.round(-2))
```

```
df = pd.DataFrame({'f': [123.456, 654.321], 'i': [123, 654], 's': ['abc', 'xyz']})

print(df)
print("\n")

print(df.dtypes)
```

```
#round()の引数を省略した場合および整数を指定した場合は全ての列が対応する桁数に丸められる

print(df.round())
print("\n")

print(df.round(2))
print("\n")

print(df.round(-2))
```

```
#キーを列名、値を桁数とする辞書を引数に指定すると、各列に個別に桁数を指定することができる。指定しない列は変化なし

print(df.round({'f': 2, 'i': -1}))
print("\n")

print(df.round({'i': -2}))
```
