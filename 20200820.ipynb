```
x = 1
for i in range(1,1000000):
  if i % 2 == 0:
    x = x + (1/(1+2*i))
  else:
    x = x - (1/(1+2*i))

print(x*4)
```

```
K = 1000000
x = 0
for i in range(1,K):
    x = x + (1/(1+2*(K-i)))*((-1)**(K-i))

print((1+x)*4)
```
