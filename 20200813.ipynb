```
#n進数への変換関数 荒木ver.
print("変換したい数を入力してね")
a=input(">>")
print("好きな自然数を入力してね")
n=input(">>")

answer=""
k=int(a)
p=int(n)

while k>=p:
  r=k%p
  k=k//p
  if k>=p:
    answer=str(r)+answer
  else:
    print(str(k)+str(r)+answer)
    break
```

```
#n進数への変換関数 小野ver.
y=input(">>")
z=input(">>")

x=0
y=int(y)
z=int(z)
m=""

for k in range(1,y):
 x=(y%(z**k))//(z**(k-1))
 y=y-x*(z**(k-1))
 m=str(x)+m
 if y==0:
  break

print(m)
```
